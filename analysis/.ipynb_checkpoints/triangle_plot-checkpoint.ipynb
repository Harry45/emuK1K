### Cosmology Part

```python
c1 = '#EEC591'
c3 = '#8B0000'
alpha = 0.7
contours = np.array([0.68,0.95])

legend_1 = mpatches.Patch(color=c1, label='EMCEE', alpha = alpha)
legend_2 = Line2D([0], [0], color=c3, linewidth=3, linestyle='-', label = 'MontePython')
legend   = [legend_1,legend_2]
```

```python
cosmology = flatchain[:,[2, 0, 1, 3, 4, 10, 11]]
```

```python
cosmo_labels = [r'$S_{8}$', 
                r'$\Omega_{\textrm{cdm}}h^{2}$',
                r'$\Omega_{\textrm{b}}h^{2}$',
                r'$n_{s}$',
                r'$h$',
                r'$A_{\textrm{bary}}$',
                r'$A_{\textrm{IA}}$']

ndim = cosmology.shape[1]
names = ["x%s"%i for i in range(ndim)]
```

```python
samples1 = MCSamples(samples=cosmology,names = names, labels = cosmo_labels, settings = settings)
```

```python
samples1_mp = MCSamples(samples=samps1_mp,names = names, labels = cosmo_labels, settings = settings)
```

```python
G = plots.getSubplotPlotter(subplot_size=3)
samples1.updateSettings({'contours': [0.68, 0.95]})
G.triangle_plot([samples1], filled = True, line_args={'lw':3, 'color':c1}, contour_colors=[c1])
G.settings.num_plot_contours = 2
G.settings.alpha_filled_add = alpha
plt.legend(handles=legend, loc = 'best',prop={'size':40}, bbox_to_anchor=(0.7, 6.5), borderaxespad=0.)
for i in range(0,7):
    for j in range(0, i+1):
        if i != j:
            ax = G.subplots[i,j]
            
            a, b = G.get_param_array(samples1_mp, ['x'+str(j), 'x'+str(i)])
            density = G.sample_analyser.get_density_grid(samples1_mp, a, b)
            density.contours = density.getContourLevels(contours)
            contour_levels = density.contours
            
            ax.contour(density.x, density.y, density.P, sorted(contour_levels), colors = c3, linewidths = 3) 
            ax.tick_params(axis = 'x', labelsize = 30, rotation=45, pad=15)
            ax.tick_params(axis = 'y', labelsize = 30, pad=15)
            ax.yaxis.label.set_size(30)
            ax.xaxis.label.set_size(30)
            ax.xaxis.labelpad = 40
            ax.yaxis.labelpad = 30
        else:
            ax = G.subplots[i,j]   
            
            dense = samples1_mp.get1DDensity('x'+str(i))
            dense.normalize(by='max')
            ax.plot(dense.x, dense.P, lw = 3, c = c3)
            
            ax.tick_params(axis = 'x', labelsize = 30, rotation=45, pad=15)
            ax.tick_params(axis = 'y', labelsize = 30, pad=15)
            ax.yaxis.label.set_size(30)
            ax.xaxis.label.set_size(30)
            ax.xaxis.labelpad = 40
            ax.yaxis.labelpad = 30
# plt.savefig('plots/cosmology_cholesky_'+fname+'.pdf', transparent = False, bbox_inches = 'tight')            
plt.show()
```

### Banana Shapes

```python
omega_cdm = flatchain[:,0]
omega_b = flatchain[:,1]
S_8 = flatchain[:,2]
h = flatchain[:,4]
omega_nu = 0.06 / 93.14

omega_m = (omega_b+omega_cdm+omega_nu) / h**2 

sigma8 = S_8 * np.sqrt(0.3/omega_m)
```

```python
samps3 = np.array([omega_m, sigma8]).T
```

```python
ndim = samps3.shape[1]
names = ["x%s"%i for i in range(ndim)]

labels = [r'$\Omega_{m}$', r'$\sigma_{8}$']

samples3 = MCSamples(samples=samps3,names = names, labels = labels, settings = settings)
```
